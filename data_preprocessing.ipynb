```python
import pandas as pd
import numpy as np
from datetime import datetime
import os
import mysql.connector 
```

```python
def space():
    print(" ")
    print("---------------------------------------------------------------")
    print(" ") 

def basic_exploration(fname):
    df = pd.read_csv(fname)
    steps_title = ["head", "shape", "duplicates if any","null and its sum", "No. of unique values across df", "Get to know datatypes"]
    steps = [df.head(2),df.shape, df.duplicated().any(), df.isnull().sum(), df.nunique(), df.dtypes]
    for i in range(len(steps)):
        print("----",steps_title[i], "-------")
        print(steps[i])
        space()


# basic_exploration("olist_products_dataset.csv")  
```
